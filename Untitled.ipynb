```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')

# Load dataset
df = pd.read_csv("Daily Household Transactions.csv")

# --- Data Cleaning ---
print("Cleaning data...")
df.dropna(subset=['Date', 'Amount'], inplace=True)  
df.fillna({'Category': 'Unknown', 'Subcategory': 'Unknown'}, inplace=True)  
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce')  
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce').astype(float)
df.drop_duplicates(inplace=True)

# Create new features
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day_name()
df['Month_Year'] = df['Date'].dt.to_period('M')
df['Transaction_Type'] = df['Income/Expense'].apply(lambda x: "Income" if x == "Income" else "Expense")

# Set global style for better aesthetics
sns.set_theme(style="whitegrid", context="notebook", font_scale=1.1)

# Define color palette
palette = {"Income": "#2ecc71", "Expense": "#e74c3c", "Unknown": "#95a5a6"}
category_palette = sns.color_palette("husl", n_colors=len(df['Category'].unique()))

# --- Exploratory Data Analysis ---
print("\nAnalyzing transaction patterns...")

# 1 Save Transaction Distribution
plt.figure(figsize=(14, 7))
sns.histplot(df['Amount'], bins=50, kde=True, color="#3498db", edgecolor='#2c3e50', alpha=0.8)
plt.title('Transaction Amount Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Amount ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)
sns.despine()
plt.tight_layout()
plt.savefig("transaction_distribution.png", dpi=300)  # Save plot
plt.show()

# 2 Save Category Distribution
plt.figure(figsize=(14, 8))
ax = sns.countplot(data=df, x='Category', order=df['Category'].value_counts().index, palette=category_palette, edgecolor='#2c3e50')
plt.title('Transaction Distribution by Category', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, fontsize=11)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Category', fontsize=12)

# Add percentage labels
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height()/total:.1f}%'
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)

sns.despine()
plt.tight_layout()
plt.savefig("category_distribution.png", dpi=300)  # Save plot
plt.show()

# 3 Save Income vs Expense Distribution
plt.figure(figsize=(10, 8))
ax = sns.countplot(data=df, x='Transaction_Type', palette=palette, edgecolor='#2c3e50', order=["Income", "Expense"])
plt.title('Income vs Expense Distribution', fontsize=16, fontweight='bold')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Transaction Type', fontsize=12)

# Add value labels
for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12, fontweight='bold')

sns.despine()
plt.tight_layout()
plt.savefig("income_vs_expense.png", dpi=300)  # Save plot
plt.show()

# --- Time Series Analysis ---
print("\nAnalyzing time-based trends...")

# Save Monthly Financial Trends
monthly_data = df.groupby('Month_Year').agg(Total_Amount=('Amount', 'sum')).reset_index()
monthly_data['Month_Year'] = monthly_data['Month_Year'].dt.to_timestamp()

plt.figure(figsize=(16, 8))
plt.plot(monthly_data['Month_Year'], monthly_data['Total_Amount'], marker='o', linestyle="-", color="#34495e", linewidth=2.5, label='Total Amount')
plt.fill_between(monthly_data['Month_Year'], monthly_data['Total_Amount'], color="#bdc3c7", alpha=0.3)
plt.title('Monthly Financial Trends', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.grid(True, linestyle="--", linewidth=0.5, alpha=0.7)
plt.axhline(0, color='black', linewidth=1)
plt.legend(loc='upper left', fontsize=12)
plt.xticks(rotation=45)
sns.despine()
plt.tight_layout()
plt.savefig("monthly_trends.png", dpi=300)  # Save plot
plt.show()

# --- Summary Report ---
print("\nGenerating summary report...")

total_income = df[df['Transaction_Type'] == 'Income']['Amount'].sum()
total_expense = df[df['Transaction_Type'] == 'Expense']['Amount'].sum()
net_savings = total_income - total_expense
savings_rate = (net_savings / total_income) * 100 if total_income > 0 else 0
top_category = df.groupby('Category')['Amount'].sum().idxmax()
avg_daily_spend = df.groupby('Day')['Amount'].mean().mean()
analysis_period = f"{df['Date'].min().date()} to {df['Date'].max().date()}"
transactions_analyzed = len(df)

# Save summary report to a text file
with open("summary_report.txt", "w", encoding="utf-8") as f:
    f.write("=" * 60 + "\n")
    f.write("FINANCIAL ANALYSIS SUMMARY".center(60) + "\n")
    f.write("=" * 60 + "\n")
    f.write(f"\n Total Income: ${total_income:,.2f}")
    f.write(f"\n Total Expenses: ${total_expense:,.2f}")
    f.write(f"\n Net Savings: ${net_savings:,.2f}")
    f.write(f"\n Savings Rate: {savings_rate:.1f}%")
    f.write(f"\n Most Spent Category: {top_category}")
    f.write(f"\n Average Daily Spending: ${avg_daily_spend:.2f}")
    f.write(f"\n Analysis Period: {analysis_period}")
    f.write(f"\n Transactions Analyzed: {transactions_analyzed:,}")
    f.write("\n" + "=" * 60)

print("\n Summary report saved successfully: `summary_report.txt`")

```
